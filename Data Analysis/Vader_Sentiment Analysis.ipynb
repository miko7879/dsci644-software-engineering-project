This sentiment analysis uses the Vader package based on the NLTK library. It achieves an 86% accuracy.

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
df_coc = pd.read_csv("AppReview.csv", encoding="utf-8")
df_coc.head()
```

```python
def calc_vader_sentiment(text):
    sentiment = "Neutral"
    vs = analyzer.polarity_scores(str(text))
    compound = vs['compound']
    
    if(compound >= 0):
        sentiment = "Positive"
    elif(compound < 0):
        sentiment = "Negative"
    return sentiment
```

```python
df = pd.DataFrame({
    'uid': list(df_coc.index.values),
    'text': df_coc['reviewText'],
    'rating': df_coc['reviewerRating'],
    'label': [calc_vader_sentiment(x) for x in df_coc['reviewText']]
})
```

```python
df.head(25)
```

```python
import numpy as np
df["P/N"] = np.where(df['rating']>0.5, "Positive", "Negative")
df["Correct"] = np.where(df['P/N'] == df["label"], True, False)
df.head(25)
```
